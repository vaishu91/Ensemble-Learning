```python
import numpy as np
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
X = np.array([2.5, 4, 9, 5.6, 7.2, 3.0, 2.1, 8, 4.9, 7.3]).reshape(-1, 1)
y = np.array([0, 1, 1, 0, 1, 0, 0, 1, 1, 1])
```

```python
dt_classifier = DecisionTreeClassifier(random_state=0)
bagging_classifier = BaggingClassifier(base_estimator=dt_classifier, n_estimators=10)
bagging_classifier.fit(X, y)
y_pred = bagging_classifier.predict(X)
accuracy = np.mean(y == y_pred)
print(f"Bagging Accuracy: {accuracy:.2f}")
```

```python
dt_classifier = DecisionTreeClassifier(random_state=0)
boosting_classifier = AdaBoostClassifier(base_estimator=dt_classifier, n_estimators=10)
boosting_classifier.fit(X, y)
y_pred = boosting_classifier.predict(X)
accuracy = np.mean(y == y_pred)
print(f"Boosting Accuracy: {accuracy:.2f}")
```

```python
x_min, x_max = X.min() - 1, X.max() + 1
xx = np.linspace(x_min, x_max, 1000).reshape(-1, 1)
```

```python
Z_bagging = bagging_classifier.predict(xx)
Z_boosting = boosting_classifier.predict(xx)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X, y, c='k', marker='x', label='Data Points')
plt.plot(xx, Z_bagging, c='b', label='Bagging Decision Boundary')
plt.xlabel('Feature')
plt.ylabel('Class')
plt.title('Bagging and Boosting Decision Boundaries')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X, y, c='k', marker='x', label='Data Points')
plt.plot(xx, Z_boosting, c='r', label='Boosting Decision Boundary')
plt.xlabel('Feature')
plt.ylabel('Class')
plt.title('Bagging and Boosting Decision Boundaries')
plt.legend(loc='best')
plt.show()
```
